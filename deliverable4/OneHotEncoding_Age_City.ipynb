```python
fact_df = pd.DataFrame(fact_table, columns=['Individual_Key', 'Resolved', 'Fatal', 'Special_Measure_Key', 'PHU_Key', 'Weather_Key', 'Onset_Date_Key', 'Reported_Date_Key', 'Test_Date_Key', 'Specimen_Date_Key'])
fact_df.head()
```

```python
individual_df = pd.DataFrame(individual_table, columns=['surrogate_key', 'individual_id', 'date', 'age_group', 'gender', 'cause_of_infection', 'outbreak_related', 'reporting_phu_city'])
individual_df.head()
```

### Age Group Label Encoding

```python
# creating instance of labelencoder
labelencoder = LabelEncoder()
# assign numerical values and store in another column
individual_df['Age_Num'] = labelencoder.fit_transform(individual_df['age_group'])
individual_df.head()
```

### Age Group One Hot Encoding

```python
encoder = OneHotEncoder(handle_unknown='ignore')
```

```python
encoder_df = pd.DataFrame(encoder.fit_transform(individual_df[['Age_Num']]).toarray())
```

```python
encoder_df['Age_Num'] = individual_df['Age_Num']
```

```python
individual_df = individual_df.join(encoder_df, on='Age_Num', how='left', lsuffix='_left', rsuffix='_right')
individual_df.rename(columns={0: "20s", 1: "30s", 2: "40s", 3: "50s", 4: "60s", 5: "70s", 6: "80s", 7: "<20", 8: "90+"},inplace = True)
individual_df.head()
```

```python
individual_df=individual_df.drop(['gender', 'cause_of_infection', 'outbreak_related', 'Age_Num_left', 'Age_Num_right', 'Age_Num_right'], axis=1)
individual_df.head()
```

### City Label Encoding

```python
individual_df['City_Num'] = labelencoder.fit_transform(individual_df['reporting_phu_city'])
individual_df.head()
```

### City One Hot Encoding

```python
encoder2_df = pd.DataFrame(encoder.fit_transform(individual_df[['City_Num']]).toarray())
```

```python
encoder2_df['City_Num'] = individual_df['City_Num']
```

```python
individual_df = individual_df.join(encoder2_df, on='City_Num', how='left', lsuffix='_left', rsuffix='_right')
individual_df.rename(columns={0: "Ottawa", 1: "Toronto"},inplace = True)
individual_df.head()
```

```python
individual_df=individual_df.drop(['age_group', 'reporting_phu_city', 'City_Num_left', 'City_Num_right'], axis=1)
individual_df.head()
```
